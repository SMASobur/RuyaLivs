```python
import requests

def send_request():
    # Articles (GET https://api.fortnox.se/3/articles)

    try:
        r = requests.get(
            url="https://api.fortnox.se/3/articles?filter=active",
            headers = {
                "Access-Token":"873fb692-e1ba-48f7-8b83-4926f0adad98",
                "Client-Secret":"BJlw5IuY1p",
                "Content-Type":"application/json",
                "Accept":"application/json",
            },
        )
        
        return r.content
    except requests.exceptions.RequestException as e:
        print('HTTP Request failed')
```

```python
json_data=send_request()
```

```python
import json
```

```python
json_data
```

```python
print(json_data)
```

```python
json_object = json.loads(json_data)
```

```python
json_formatted_str = json.dumps(json_object, indent=2)
```

```python
print(json_formatted_str.decode("utf-16") )
```
